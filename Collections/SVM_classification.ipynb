## Three halfmoon dataset

```python
# create three half moon
import numpy as np
import math
def halfmoon_v2(rad, width, dis, n_sample):

    data = np.random.rand(n_sample, 2) # create two half moon data
    labeledClass = np.zeros((n_sample, 1)) # include class 1, class 2, and class 3
    
    # 0 ~ 1/3, 1/3 ~ 2/3, 2/3 ~ 1 
    
    for i in range(0, int(n_sample / 2)):
        radius = (rad - width / 2) + (width * np.random.rand())
        theta = np.pi * np.random.rand() # rad
        data[i, 0] = radius * math.cos(theta)
        data[i, 1] = radius * math.sin(theta)
        labeledClass[i, 0] = 1 # class 1
    
    for i in range(int(n_sample / 2), int((3 * n_sample) / 4)):
        radius = (rad - width ) + (width * np.random.rand())
        theta = np.pi * np.random.rand() # rad
        data[i, 0] = radius * math.cos(-theta) + rad
        data[i, 1] = radius * math.sin(-theta) - dis
        labeledClass[i, 0] = 2 # class 2
    
    for i in range(int((3 * n_sample) / 4), int(n_sample)):
        radius = (rad - width ) + (width * np.random.rand())
        theta = np.pi * np.random.rand()
        data[i, 0]  = radius * math.cos(-theta) - rad # only difference
        data[i, 1] = radius * math.sin(-theta) - dis
        labeledClass[i, 0] = 0 # class 3
        
    return (data, labeledClass)

# Draw three half moons
N = 2000
Radius = 12
distance = -5
width = 6
(data_3halfmoon, TargetClass) = halfmoon_v2(Radius, width, distance, N)

# stack data
data_merge = np.hstack((data_3halfmoon, TargetClass))
print(data_merge.shape)

import matplotlib.pyplot as plt

plt.scatter(data_3halfmoon[0: int(N / 2) - 1, 0], data_3halfmoon[0: int(N / 2) - 1, 1], color='red', marker = '*')
plt.scatter(data_3halfmoon[int(N / 2): int((3 * N) / 4), 0], data_3halfmoon[int(N / 2): int((3 * N) / 4), 1], color = 'blue', marker = 'o')
plt.scatter(data_3halfmoon[int((3 * N) / 4): int(N), 0], data_3halfmoon[int((3 * N) / 4): int(N), 1], color = 'yellow', marker = 'o')
plt.show()
```

## Two halfmoon dataset

```python
# create two half moon
import numpy as np
import math
def halfmoon(rad, width, dis, n_sample):

    data = np.random.rand(n_sample, 2) # create two half moon data
    labeledClass = np.zeros((n_sample, 1)) # include class 1 and class 2
    
    for i in range(0, int(n_sample / 2)):
        radius = (rad - width / 2) + (width * np.random.rand())
        theta = np.pi * np.random.rand() # rad
        data[i, 0] = radius * math.cos(theta)
        data[i, 1] = radius * math.sin(theta)
        labeledClass[i, 0] = 1
    
    for i in range(int(n_sample / 2), int(n_sample)):
        radius = (rad - width / 2) + (width * np.random.rand())
        theta = np.pi * np.random.rand() # rad
        data[i, 0] = radius * math.cos(-theta) + rad
        data[i, 1] = radius * math.sin(-theta) - dis
        labeledClass[i, 0] = -1
        
    return (data, labeledClass)
    
```

## SVM classification - binary classification

```python
from matplotlib import style
from sklearn.svm import SVC

# Plot decision boundaries, vectors, etc.
def plotFunction(clf, title):
    
    ax = plt.gca()
    
    # plot the decision function
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    print(xlim, ylim)

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30) # 30
    yy = np.linspace(ylim[0], ylim[1], 30) # 30
    YY, XX = np.meshgrid(yy, xx)
    print("XX: ", XX.shape, "YY: ", YY.shape)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    print("xy: ", xy.shape)
    Z = clf.decision_function(xy)#.reshape(XX.shape)
    print("Z: ", Z.shape)
    Z = Z.reshape(XX.shape)
    print("Z(reshape): ", Z.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='r', levels=[-1, 0, 1], alpha=0.5, 
               linestyles=['-', '--', '-'])
    ax.contourf(XX, YY, Z)

    # plot support vectors
    ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
               linewidth=1, facecolors='none', edgecolors='k')
    ax.set_title(title)

clf = SVC(kernel='rbf',gamma='auto', C=0.01)
clf.fit(data, labeledClass)
print(clf)
plt.figure(1)
plt.scatter(data[:, 0], data[:, 1], cmap=plt.cm.Paired) # s = 20
plotFunction(clf, 'SVM with RBF kernel(C=0.01)')

clf1 = SVC(kernel='rbf', gamma='auto', C=1)
clf1.fit(data, labeledClass)
plt.figure(2)
plt.scatter(data[:, 0], data[:, 1], cmap=plt.cm.Paired) # s = 20
plotFunction(clf1, 'SVM with RBF kernel(C=1)')

clf2 = SVC(kernel='rbf', gamma='auto', C=100)
clf2.fit(data, labeledClass)
plt.figure(3)
plt.scatter(data[:, 0], data[:, 1], cmap=plt.cm.Paired) # s = 20
plotFunction(clf2, 'SVM with RBF kernel(C=100)')

plt.show()  
```

## Evaluate for SVM binary classes

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(data, labeledClass, test_size=0.3) # 70% training and 30% test

from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='rbf', gamma='auto', C=1) # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred))
```
